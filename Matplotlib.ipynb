# Matplotlib
Matplotlib is a python library used for data visualization. You can create bar-plots, scatter-plots,histo-grams and lot more with "Matplotlib".


```python
from numpy import *
import matplotlib.pyplot as plt

first=arange(1,11)     #  x-axis
second=2*first         #  y-axis

print(first)
print("\n")
print(second)
plt.plot(first,second )   #  line plot  just displays lists
#  here "first"  is  x-axis
#  "second" is y-axis
plt.show()       # displays the graph
```

```python
x=arange(1,10)
print(x)
print("Type of x is\t",type(x))
y=x*2
print(y)
print("Type of y is\t",type(y))

plt.plot(x,y,color='r',linestyle=':',linewidth=2)
plt.title("Data Visualisation")
plt.xlabel("x axis data")
plt.ylabel("y axis data")
plt.grid(True)
plt.show()
```

```python
years=arange(2016,2022)
print("years is\t",years)
dac_batch=array([100,110,120,115,118,120])
dbda_batch=array([38,45,48,50,55,60])
plt.plot(years,dac_batch,color='r',linestyle='-',linewidth=2)
plt.plot(years,dbda_batch,color='blue',linestyle='-',linewidth=2)
plt.title("VITA DAC_DBDA Batch Summary for last 5 years")
plt.xlabel("years")
plt.ylabel("DAC and DBDA Total no. of students")
plt.grid(True)
plt.show()
```

```python
first=arange(1,11)
second=2*first

plt.scatter(first,second)     #  scatter plot  ,   first and second must have same no. of elements 
plt.show()
```

```python
first=arange(1,11)
second=2*first

plt.scatter(first,second,marker=".",c='r',s=200)     #  scatter plot  ,   first and second must have same no. of elements 
plt.xticks(first)
plt.yticks(second)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define data
years = np.arange(2016, 2022)
dac_batch = np.array([100, 110, 120, 115, 118, 120])
dbda_batch = np.array([38, 45, 48, 50, 55, 60])

# Scatter plots for DAC and DBDA batch
plt.scatter(years, dac_batch, marker=".", c='r', s=200, label="DAC Batch")
plt.scatter(years, dbda_batch, marker=".", c='blue', s=200, label="DBDA Batch")

# Adding title and labels
plt.title("VITA DAC_DBDA Batch Summary for Last 5 Years")
plt.xlabel("Years")
plt.ylabel("DAC and DBDA Total No. of Students")

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

### Box Plot 
It displays the five-number summary of a set of data. 
The five-number summary is the minimum, first quartile, median, third quartile, and maximum.

The first quartile (Q1), is defined as the middle number between the smallest number and the median of the data set, the second quartile (Q2)  median of the given data set while the third quartile (Q3), is the middle number between the median and the largest value of the data set.


```python
import statistics
first=[1,2,3,4,5,6,7,8,9,15]
second=[1,2,3,4,5,4,3,2,1]
third=[6,7,8,9,8,7,6,5,4]

print("Median of first is\t",statistics.median(first))
print("Median of second is\t",statistics.median(second))
print("Median of third is\t",statistics.median(third))

data=list([first,second,third])
plt.boxplot(data)
plt.show()
```

```python
first=[1,2,3,4,5,6,7,8,9]
second=[1,2,3,4,5,4,3,2,1]
third=[6,7,8,9,8,7,6,5,4]

data=list([first,second,third])
plt.violinplot(data,showmedians=True,quantiles=[[0.25,0.5,0.75],[0.25,0.5,0.75],[0.25,0.5,0.75]])
#plt.violinplot(data)
#print(plt.violinplot.__doc__)
print(statistics.mean(first))
print(statistics.mean(second))
print(statistics.mean(third))
plt.show()
```
