##### Example 10

#### How to download images from a particular URL using BeautifulSoup

 - tqdm can help you create progress bars for data processing, training machine learning models, 
multi-loop Python function, and downloading data from the internet. 

 - The rfind() method returns the highest index of the specified substring (the last occurrence of the substring) in the given string.

```python
from tqdm import tqdm
r=requests.get("https://www.gsmarena.com/vivo_x200_pro_mini_handson_review-news-64924.php")

# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')
 
imgs=soup.find_all("img")

directory='c:/temp'
for img in tqdm(imgs):
    # img.attrs is a dictionary with 'src' as a key and the image link is a value
    imglink=img.attrs.get("src")
    # to get the binary form of an image
    image=requests.get(imglink).content  # raise MissingSchema(
#requests.exceptions.MissingSchema: Invalid URL 'images/sachintravel-logo.png': No scheme supplied.
# it's because our image paths don't begin with "http or https"
    filename=directory+imglink[imglink.rfind('/'):]
    with open(filename,"wb") as f:
       f.write(image) 
```

The rfind() method returns the highest index of the specified substring (the last occurrence of the substring) in the given string.

suppose imglink is like this  "//imworld.rediff.com/worldrediff/pix/bluerarrow.png"
imglink.rfind('/') will return index of last forward slash and we will slice from that onwards so we only get filename(/bluearrow.png) of img not the whole url

##### Final Example

```python
# keep "my.html" inside the same folder where your python project is running

import requests
from bs4 import BeautifulSoup

# Load the HTML file
with open('my.html', 'r') as file:
    html_content = file.read()

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Scrape the title
title = soup.title.string
print(f"Title: {title}")

# Scrape the main heading
main_heading = soup.h1.string
print(f"Main Heading: {main_heading}")

# Scrape all paragraphs
paragraphs = soup.find_all('p')
for i, p in enumerate(paragraphs, start=1):
    print(f"Paragraph {i}: {p.text}")

# Scrape all links
links = soup.find_all('a')
for link in links:
    print(f"Link text: {link.text}, URL: {link['href']}")

# Scrape list items
list_items = soup.find_all('li')
for i, item in enumerate(list_items, start=1):
    print(f"List Item {i}: {item.text}")
```

# URL Parsing

```python
# import library
import re

# url link
myurl = "https://imsports.rediff.com:8080/score/in_asa_1083661.html"

# Protocol
protocol = re.findall('(\w+)://', myurl)
print("Protocol is\t", protocol)

# Hostname
hostname = re.findall('://([\w\.]+)', myurl)
print("Hostname is\t", hostname)

portno = re.findall(':(\d+)', myurl)
print("Port no. is\t", portno)

path = re.findall('/(\w+)/', myurl)
print("Path is\t", path)

# The $sign is used to match an expression to its
# left at the end of a string.
# t$ is an expression match to a string which ends
# with t such as "dot", "mat", "cat", "rat", etc.

filename = re.findall('(\w+\.\w+)$', myurl)
print("filename is\t", filename)
```

```python
from urllib.parse import urlparse,parse_qs


url = "https://imsports.rediff.com:8080/score/in_asa_1083661.html?subject=Python&test=practical"
parsed_url = urlparse(url)
print(parsed_url.port)
print(parsed_url.hostname)
print(parsed_url.path)
print(parse_qs(parsed_url.query))
```
